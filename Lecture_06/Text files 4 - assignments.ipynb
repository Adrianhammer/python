```c#

# open file
fname = 'babynames2000s.txt'
infile = open(fname, 'r')

# read lines
lines = infile.readlines()

# close file
infile.close()

# create empty lists
boys = []
girls = []

# for each line...
for line in lines:
    #...split the line on space
    parts = line.split()
    #...and append the boy name to boys
    boys.append(parts[1].upper())
    #...and append the girl name to girls
    girls.append(parts[3].upper())
    

# open file to save list with boy names
outfile = open('boynames.txt', 'w')

# for each name in boys...
for name in boys:
    # ...write name and newline character to file
    outfile.write(name + '\n')
outfile.close()


# open file to save list with girl names
outfile = open('girlnames.txt', 'w')
# for each name in girls...
for name in girls:
    # ...write name and newline character to file
    outfile.write(name + '\n')
outfile.close()

    
```

</p>
</details> 

### Solution 2

<details>
    
<summary> Click to expand!</summary>
<p> 

```c#

# open, read and close file
fname = 'babynames2000s.txt'
infile = open(fname, 'r')
lines = infile.readlines()
infile.close()


# define limit
LIMIT = 150000


# create empty lists
boy_list = []
girl_list = []


# for line in file
for line in lines:
    # split line and extract relevant data fields
    parts = line.split()
    rank = parts[0]
    boyName = parts[1]
    boyCount = int(parts[2].replace(',', ''))
    girlName = parts[3]
    girlCount = int(parts[4].replace(',', ''))
  
    # if boyCount exceeds limit...
    if boyCount > LIMIT:
        #...append boyName and boyCount to list as a tuple
        boy_list.append((boyName, boyCount))
    
    # if girlCount exceeds limit...  
    if girlCount > LIMIT:
        #...append girlName and girlCount to list as a tuple
        girl_list.append((girlName, girlCount))
        
        
# display result

print('--------------------')
print('GIRLNAME    : TOTAL ')
print('--------------------')

for name in girl_list:
    print(f'{name[0]:11} : {name[1]}')


print('\n--------------------')
print('BOYNAME     : TOTAL ')
print('--------------------')

for name in boy_list:
    print(f'{name[0]:11} : {name[1]}')
    
```
