<p> Write a function called <code>tempConversion</code> that converts temperatures from Fahrenheit to Celsius and vice versa. The function takes two inputs: a temperature to be converted, and a variable indicating whether the temperature is in Fahrenheit or Celsius. 
    
    
Conversion from Celsius to Fahrenheit follows the following formula: Celsius = (5 / 9) * (Fahrenheit  - 32)

    
Remember proper documentation of the function.

```c#
    
def tempConversion(which, temp):    
    """
    This function converts temperatures Fahrenheit/Celsius, and returns the
    converted temperature.

    Parameters
    ----------
    which : str
        Takes the value 'F' for Fahrenheit-Celsius conversion and the value 'C'
        for Celsius-Fahrenheit conversion..
    temp : float or int
        The temperature to be converted..

    Returns
    -------
    converted_temp : float
        The converted temperature.

    """
    
    # if conversion is F --> C
    if which == 'F':
        converted_temp = (temp-32)*5/9

    # else, conversion is C --> F
    else:
        converted_temp = (9/5)*temp+32

    return converted_temp

which = 'F'
temp = 32
converted_temp = tempConversion(which, temp)

if which == 'F':
    print('\n{temp} degree Fahrenheit equals {converted_temp:.1f} degree Celsius.')
else:
    print('\n{temp} degree Celsius equals {converted_temp:.1f} degree Fahrenheit.')
```
