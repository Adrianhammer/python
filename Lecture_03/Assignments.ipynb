# Display program welcome
print(f'This program will convert temperatures (Fahrenheit/Celsius)')

print(f'Enter C to convert from Celsius to Fahrenheit')

# Initialize Boolean flag
valid_input = False 

# Ask user for choice
which = input('Enter selection: ')
while not valid_input: # loop will execute as long as valid_input = False
    if which == 'F' or which == 'C': # if the input is valid, change the value of the Boolean flag to true
        valid_input = True
    else:
        print(f'INVALID INPUT. Only F or C is accepted.') # else, continue asking for input
        which = input('Enter selection: ')

# Ask user for temperature
temp = float(input('Enter temperature to convert: '))

# Determine temperature conversion needed and display the result
if which == 'F':
    converted_temp = (temp-32)*5/9
