    <p>   Use the DataFrame <code>titanic3</code>:
    

1. How many missing values are there for the Age column? How many of the missing observations are women, and how many are men?

    
2. Create a figure with two subplots:
    - In the first subplot, create a bar plot showing the probability of surviving by Pclass for women.
    - In the second subplot, create a bar plot showing the probability of surviving by Pclass for men. 


    
</div>

### Solution 1

<details>
    
<summary> Click to expand!</summary>
<p> 

```c#
import pandas as pd

# read data file
titanic3 = pd.read_csv('titanic3.csv')
    
# number of missing observations in Age column
titanic3.isnull().sum()['Age']
# isnull() is the same as isna()

# keep only rows with NaN in Age column 
age_missing = titanic3[titanic3['Age'].isna()]

# count the number of women and men
age_missing['Sex'].value_counts()
    
```

</p>
</details> 

### Solution 2

<details>
    
<summary> Click to expand!</summary>
<p> 

```c#
 
import matplotlib.pyplot as plt    
    
### Calculate mean survival by sex and Pclass

# split rows into women and men
women = titanic3[titanic3['Sex'] == 'female']
men = titanic3[titanic3['Sex'] == 'male']

# calculate mean survival by Pclass
women_pclass = women.groupby('Pclass')['Survived'].mean()
men_pclass = men.groupby('Pclass')['Survived'].mean()


### Create bar plots
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 4))

# plot
ax[0].bar(['1st class', '2nd class', '3rd class'], women_pclass)
ax[1].bar(['1st class', '2nd class', '3rd class'], men_pclass)

# add label and titles
ax[0].set_ylabel('Survived')
ax[0].set_title('Female')
ax[1].set_title('Male')

# set axis range
ax[0].set_ylim(0, 1)
ax[1].set_ylim(0, 1)

plt.show()
    
```
